```python
import numpy as np

n = int(input("Nhập n = "))
a = np.array(list(map(float, input("Nhập mảng a = ").split()))[:n])
b = np.array(list(map(float, input("Nhập mảng b = ").split()))[:n])
print(f"Mảng a và b: \n\t{a}\n\t{b}")

c = a + b
d = a - b
e = a * b
f = a / b
print(f"c = a + b = {c}")
print(f"d = a - b = {d}")
print(f"e = a * b = {e}")
print(f"f = a / b = {f}")

print(f"Tổng các phần tử của c: {sum(c)}")
print(f"Giá trị lớn nhất của c: {np.amax(c)}")
print(f"Giá trị nhỏ nhất của c: {np.amin(c)}")

print(f"Các phần tử chỉ số chẵn của c: {c[::2]}")
print(f"Tổng các phần tử đó là: {sum(c[::2])}")

try:
    k = int(input("Nhập số dòng k: "))
    if n % k:
        raise ValueError
    c = c.reshape(k, n // k)
    print(f"Mảng c mới: \n\t{c}")
except:
    print(f"{ValueError}: Không thể reshape c!")
```

```python
import numpy as np

n = int(input("Nhập số lượng phần tử n = "))
a = np.array(list(map(float, input("Nhập mảng a = ").split()))[:n])
try:
    k = int(input("Nhập số dòng k: "))
    if n % k:
        raise ValueError
    a = a.reshape(k, n // k)
    print(f"Ma trận a: \n{a}")
except:
    print(f"{ValueError}: Không thể chuyển c thành ma trận!")

b, c = a[:, 0], a[:, 1]
print(f"Mảng b: {b}")
print(f"Mảng c: {c}")

d = np.concatenate((b, c))
print(f"Mảng d: {d}")

print(f"Các vị trí có ptử > 1: {np.where(d > 1)}")

d = np.sort(d, kind="heapsort")
print(f"Mảng d đã sắp: {d}")


k = float(input("Nhập k = "))
p = np.searchsorted(d, k)
print(f"Vị trí cần chèn: {p}")
print(f"Mảng d sau chèn: {np.insert(d, p, k)}")
```

```python
import numpy as np


def read(file):
    try:
        with open(file, "r") as f:
            row, col = f.readline().split()
            return int(row), int(col), np.array([line.split() for line in f])
    except:
        print("Không tìm thấy file!")
        return


# 1
n, m, mx = read("./DATA54.txt")

# 2
x, y = mx[:, :-1], np.array(mx[:, -1], dtype=int)

# 3
unq, cnt = np.unique(y, return_counts=True)
# print([f"{i}: {j}" for i, j in zip(unq, cnt)])

# 4
tb = []
for i in range(len(x[0])):
    count = s = 0.0
    for j in x[:, i]:
        s += float(j) if j != "?" else 0
        count += 1 if j == "?" else 0
    tb.append(s / (len(x) - count))

# 5
for i in range(len(x[0])):
    for j in range(len(x)):
        if x[j, i] == "?":
            x[j, i] = tb[i]

# 6
def w_file(file, x):
    with open(file, "w") as f:
        for i in x:
            if isinstance(i, int):
                f.write(" ".join(str(i)) + "\n")
            elif isinstance(i, np.ndarray):
                f.write(" ".join(i) + "\n")


w_file("./X.out", x)
w_file("./Y.out", y)
# 7
pos = int(len(x) * 0.7) + 1
res_pos = len(x) - pos
x_train, x_test, y_train, y_test = [], [], [], []
for i in range(res_pos):
    idx = np.random.randint(0, len(x))
    x_test.append(x[idx, :])
    y_test.append(y[idx])
    x = np.delete(x, pos, 0)
    y = np.delete(y, pos, 0)
    x_train = np.array(x)
    y_train = np.array(y)

w_file("./x_train.out", x_train)
w_file("./x_test.out", x_test)
w_file("./y_train.out", y_train)
w_file("./y_test.out", y_test)

# 8
# print(f"Dim of x: {np.ndim(x)}")
# print(f"Size of x: {np.size(x, axis=0)}x{np.size(x, axis=1)}")
# print(x.flatten())
# print(f"x has {len(x)} elements")
```
