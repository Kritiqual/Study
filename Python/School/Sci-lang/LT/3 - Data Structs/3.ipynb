```python
n = int(input("Nhap so luong phan tu: "))
a = set(input(f"Nhap phan tu thu {i + 1}") for i in range(n))
b = set(input(f"Nhap phan tu thu {i + 1}") for i in range(n))

print(a | b)
print(a & b)
print(a - b)

print("a la con cua b" if a.issubset(b) else "a khong la con cua b")

print("a co chua set b" if a.issuperset(b) else "a khong chua set b")

c = input("Nhap gia tri can tim: ")
if c in a:
    print(f"{c} co trong set a")
elif c in b:
    print(f"{c} co trong set b")
else:
    print(f"{c} khong co trong a va b")
```

```python
str = input("Nhap string: ")
print(f"String co {len(str.split())} tu")


d = {i: str.split()[i] for i in range(len(str.split()))}
print(d)


def ck(s):
    a = b = 0
    for i in range(len(s)):
        if s[i] == "(":
            a += 1
        elif s[i] == ")":
            b += 1
        if a < b:
            return False
    if a != b:
        return False
    return True


print("YES" if ck(input("Nhap bieu thuc: ")) else "NO")
```
