```python
import numpy as np


# Đọc file
def read(path):
    return np.loadtxt(path, dtype='str', delimiter='\t')


# Tách mảng thành 2 mảng:
# Mảng label gồm 1 cột cuối
# Mảng value gồm các cột còn lại (trừ 4 cột đầu) đã được chuyển sang kiểu float
def array_split(array):
    value, labels = np.hsplit(array[:, 4:], [-1])
    return value.astype('float32'), labels


# Tính khoảng cách Euclidean giữa hai dòng bất kỳ được truyền vào
def distance(row1, row2):
    return ((row1 - row2)**2).sum()**0.5


# Tính khoảng cách Euclidean giữa 1 dòng với tất cả các dòng trong mảng
def distance2all(row, x):
    return np.array([distance(row, i) for i in x])


# Trả về label dự đoán
def predict(test_row, values, labels):
    dis = distance2all(test_row, values)
    index = dis.argmin()
    return labels[index]


if __name__ == '__main__':
    # Đường dẫn tới các file:
    data_path = './kddcup.data'
    test_path = './kddcup.test'

    # [1] Đọc toàn bộ tệp kddcup.data lên mảng x_train;
    # cắt bỏ 4 cột đầu tiên và cắt cột cuối cùng của x_train ra một mảng label;
    # chuyển x_train về kiểu float.
    x_train = read(data_path)
    x_train, labels = array_split(x_train)

    # [2] Đọc toàn bộ file kddcup.test lên mảng x_test;
    # cắt bỏ 4 cột đầu và cột cuối cùng của x_test rồi chuyển mảng về kiểu thực.
    x_test = read(test_path)
    x_test, test_labels = array_split(x_test)

    # [3] Lấy 1 dòng (đặt tên là test) bất kỳ từ mảng x_test hãy dự đoán label của nó bằng cách:
    # Tính khoảng cách từ nó tới tất cả các dòng của x_train.
    # Label của test là label của dòng gần với nó nhất.
    np.random.seed(29082002)
    ran = np.random.randint(0, len(x_test))
    test = x_test[ran]
    label = predict(test, x_train, labels)
    print(f'label dự đoán   = {label}')
    print(f'label chính xác = {test_labels[ran]}')

```

```python
import numpy as np
def readfile(filename):    
    try:
        with open(filename, mode = 'r') as f:
            x = []
            while True:
                line = f.readline().split()
                if not line:
                    break
                x.append(line)
            return x    
    except:
        print("BUG")
        return



readfile("./kddcup.data1")
```
