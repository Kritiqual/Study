```python
# Bai 1
import os

championLOL = ["Yasuo", "Lee Sin", "Zed", "Master Yi", "Garen", "Tryndamere"]
dataLOL = [
    {"price": 6300, "Ulti": "Trăn trối"},
    {"price": 4800, "Ulti": "Nộ Long Cước"},
    {"price": 4800, "Ulti": "Dấu Ấn Tử Thần"},
    {"price": 450, "Ulti": "Chiến Binh Sơn Cước"},
    {"price": 450, "Ulti": "Công Lý Demacia"},
    {"price": 1350, "Ulti": "Từ Chối Tử Thần"},
]
D = dict(
    sorted(
        dict(zip(championLOL, dataLOL)).items(),
        key=lambda x: x[1]["price"],
        reverse=True,
    )
)
min_Price = int(input("Nhap Min Price: "))
while min_Price < 0:
    os.system("clear")
    min_Price = int(input("Nhap Min Price: "))
max_Price = int(input("Nhap Max Price: "))
while max_Price < min_Price:
    os.system("cls")
    print("Min Price:", min_Price)
    max_Price = int(input("Nhap Max Price: "))
print({i: D.get(i) for i in D if min_Price <= D[i].get("price") <= max_Price})
```

```python
# Bai 2
import os, random

n = int(input("Nhap so luong tai khoan: "))
while not 10 <= n <= 100000:
    os.system("cls")
    n = int(input("Nhap so luong tai khoan: "))
Username = 2020600000
D = {}
for i in range(n):
    Username += 1
    D["Account" + str(i + 1)] = {
        "Username": str(Username),
        "Password": random.choice(["CNTT", "KHMT", "KTPM", "HTTT"]) + str(Username),
    }
print(D)
```
